```python
from transformers import BartTokenizer, BartForConditionalGeneration
```

```python
tokenizer = BartTokenizer.from_pretrained('facebook/bart-large-cnn')
model = BartForConditionalGeneration.from_pretrained('facebook/bart-large-cnn')
```

```python
text = """
The history of aviation began in the 18th century with the invention of the hot air balloon. Over time, powered flight was developed, leading to the Wright brothers’ success in 1903. Airplanes rapidly evolved, especially during wartime, pushing innovation forward. Today, aviation plays a crucial role in global transportation, commerce, and communication. Modern aircraft are faster, safer, and more efficient than ever before. As technology advances, the future of flight may include electric planes, autonomous systems, and even commercial space travel.
"""
```

```python

inputs = tokenizer([text], max_length=1024, return_tensors='pt', truncation=True)

summary_ids = model.generate(inputs['input_ids'], max_length=130, min_length=30, length_penalty=2.0, num_beams=4, early_stopping=True)

summary = tokenizer.decode(summary_ids[0], skip_special_tokens=True)
print("Resumo:\n", summary)
```
